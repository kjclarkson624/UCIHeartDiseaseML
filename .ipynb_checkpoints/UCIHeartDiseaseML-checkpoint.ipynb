```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
uci = pd.read_csv('heart.csv')
uci
```

age: age in years
sex: (1 = male; 0 = female)
cpchest: pain type
trestbps: resting blood pressure (in mm Hg on admission to the hospital)
cholserum: cholestoral in mg/dl
fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
restecg: resting electrocardiographic results
thalach: maximum heart rate achieved
exang: exercise induced angina (1 = yes; 0 = no)
oldpeak: ST depression induced by exercise relative to rest
slope: the slope of the peak exercise ST segment
ca: number of major vessels (0-3) colored by flourosopy
thal: 3 = normal; 6 = fixed defect; 7 = reversable defect
target: 1 or 0

## Data Visualization

```python
sns.countplot(x="target", data=uci)
```

```python
sns.countplot(x="sex", data=uci)
```

```python
## most of our observations had no chest pain.
## compare cp with target, see how many have what

sns.countplot(x="cp", data=uci)
```

```python
sns.distplot((uci['age']), kde=False, rug=True);
```

```python
sns.distplot((uci['trestbps']), kde=False, rug=True);
```

```python
## a good chunk of our observations here are between 200 and 300 chol
## we have one value all the way out at the high range of 500, maybe we can see if that person has heart disease?
sns.distplot(uci['chol'], kde=False, rug=True);
```

```python
## plot stats for male vs female chol
## female seem to have on average higher chol

sns.catplot(data = uci, x = 'sex', y = 'chol', kind='box')
```

```python
## indeed, the person with highest cholestoral has heart disease.
## interesting to note that women in this dataset have heart disease at every range.
sns.catplot(data = uci, x = 'sex', y = 'chol', hue = 'target',
            kind = "swarm")
```

```python
# Youngest patient (Man aged 29) has heart disease while oldest patient (Man aged 77) does not. 
# Majority of women without heart disease are between the ages of 55 and 64 

sns.catplot(data = uci, x = 'sex', y = 'age', hue = 'target',
            kind = "swarm")
```

```python
## let's investigate the relationship between our target, age, sex, and cholestorol
uci2 = uci[['target', 'age', 'sex', 'chol']]
sns.pairplot(uci2, hue='target', height=2.5)
```

```python
corr = uci.corr()
g = sns.heatmap(corr,  vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, fmt='.2f', cmap='coolwarm')
sns.despine()
g.figure.set_size_inches(20,20)
    
plt.show()
```
